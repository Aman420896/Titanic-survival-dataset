Data Preprocessing

Replacing values with with integers

```python
df['Sex'].replace({'male' : 0,'female' : 1}, inplace = True)
```

```python
df.head()
```

```python
df['Embarked'].replace({'S' : 0,'C' : 1,'Q' : 2}, inplace = True)
```

```python
df.head()
```

Dropping the least valueable columns

```python
df = df.drop(columns = ['Name','Ticket','Age'],axis = 1)
```

```python
df.head()
```

Differeniating independent and dependent variables

```python
x = df.drop(columns = ['Survived'],axis = 1)
y = df['Survived']
```

Splitting the data for  training  and testing the model

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state = 3)
```

```python
x.shape
```

```python
print(x_train.shape,y_train.shape,x_test.shape)
```

```python
x.head()
```

Trainig the dataset through LogisticRegression Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

Fitting the data into the model

```python
model.fit(x_train,y_train)
```

```python
x.isnull().sum()
```

Predicting the values

```python
x_train_prediction = model.predict(x_train)
```

Getting the accuracy score

```python
from sklearn.metrics import accuracy_score

x_train_accuracy = accuracy_score(x_train_prediction,y_train)
print(x_train_accuracy)
```

Predicting the test data

```python
x_train_prediction = model.predict(x_test)
x_test_accuracy = accuracy_score(x_train_prediction,y_test)
print(x_test_accuracy)
```
